```python
#Importing Pandas and Matplotlib and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
sns.set(style="ticks", color_codes=True)
```

```python
df= pd.read_csv('Data/dataset_caps.csv')
```

```python
#Printing dataset head
df.head()
```

```python
#Printing datset tail
df.tail()
```

```python
#Plotting the Data 
df.info()
```

```python
#Histogram Plot with Density Curve
sns.distplot(df.age, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel ('Age')
plt.ylabel ('Distribution')
plt.savefig('pic2.png')
```

```python
#Scatter plot of different variables in order to have insight into correlation
```

```python
sns.regplot(x="scl", y="rbp", data=df[df.scl!=0], fit_reg=True, color='g')
plt.xlabel ('Serum Cholesterol Level')
plt.ylabel ('Resting Blood Pressure')
plt.title ('Relationship between Cholesterol Level and Resting Blood Pressure')

#Note we have some extreme values, however, there seems to be a positive correlation between both variables. 
```

```python
sns.swarmplot(x='eia', y='mhra', data=df)
plt.xlabel('Exercise Induced Angina')
plt.ylabel('Maximum Heart Rate Achieved')
plt.title('Exercise Induced Angina and Maximum Heart Rate Achieved')
```

```python
sns.boxplot(x="new_dhd", y="mhra", data=df)
plt.xlabel('Greater/Less than 50% Blockage')
plt.ylabel ('Maximum Heart Rate Achieved')
label1=('<50%', '>50%')
plt.xticks(np.arange(len(label1)), label1)

```

```python
sns.swarmplot(x='sex', y='rbp', data=df)
plt.xlabel('Gender')
plt.ylabel('Resting Blood Pressure')
plt.title('Gender and Resting Blood Pressure')
```

```python
sns.boxplot(x='new_dhd', y='rbp', data=df)
plt.xlabel('Greater/Less than 50% Blockage')
plt.ylabel('Resting Blood Pressure')
label=('<50%', '>50%')
plt.xticks(np.arange(len(label)), label)
```

```python
objects5=('Typical Angina Pain', 'Atypical Angina Pain', 'Non-Anginal Pain', 'Asymptomatic Pain')
sns.boxplot(x='cp', y='mhra', data=df)
plt.xticks(np.arange(len(objects5)), objects5, rotation=45)

plt.xlabel('Chest Pain Type')
plt.ylabel('Maximum Heart Rate Achieved')
plt.title('Relationship between MHRA and Chest Pain')
```

```python
sns.swarmplot(x='cp', y='rbp', data=df)
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Pressure')
plt.title('Chest Pain Type and Resting Blood Pressure')
```

```python
objects6=('Normal', 'ST Wave Abnormality', 'Left Ventricular Hypertrophy')
sns.swarmplot(x='rer', y='rbp', data=df)
plt.xticks(np.arange(len(objects6)), objects6, rotation=25)
plt.xlabel('Resting Electrocardiographic Results')
plt.ylabel('Resting Blood Pressure')
plt.title('Resting Electrocardiographic Results and Resting Blood Pressure')
```

```python
#Slicing dataset by gender
male= df[df.sex==1]
female= df[df.sex!=1]
```

```python
male.rbp.describe()
```

```python
#Histogram Plot for males and females resting blood pressure

plt.subplot(2,1,1)
plt.hist(male.rbp.dropna(), color='green')
plt.xlabel('Resting Blood Pressure (Male)')
plt.ylabel('Distribution')
plt.xlim([0,200])
plt.subplot(2,1,2)
plt.hist(female.rbp.dropna(), color='orange')
plt.xlabel('Resting Blood Pressure (Female)')
plt.ylabel('Distribution')
plt.xlim(0,200)
plt.tight_layout()
plt.show()
plt.savefig('pic2.png')

```

```python
#Histogram plot for males and female serum cholesterol levels
plt.subplot(2,1,1)
plt.hist(male.scl.dropna(), color='green')
plt.xlabel('Serum Cholesterol Level (Male)')
plt.ylabel('Distribution')
plt.xlim([80,500])
plt.subplot(2,1,2)
plt.hist(female.scl.dropna(), color='orange')
plt.xlabel('Serum Cholesterol Level (Female)')
plt.ylabel('Distribution')
plt.tight_layout()
plt.xlim([80,500])

plt.show()
plt.savefig('pic3.png')
```
