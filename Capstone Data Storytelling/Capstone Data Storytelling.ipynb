```python
#Importing Pandas and Matplotlib and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
sns.set(style="ticks", color_codes=True)
```

```python
df= pd.read_csv('Data/dataset_caps.csv')
```

```python
#Printing dataset head
print(df.head())
```

```python
#Printing datset tail
print(df.tail())
```

```python
#Plotting the Data 
df.info()
```

```python
#Histogram Plot with Density Curve
sns.distplot(df.age, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel ('Age')
plt.ylabel ('Distribution')
```

```python
#Scatter plot of different variables in order to have insight into correlation
```

```python
plt.scatter(x="scl", y="rbp", data=df, s=15)
plt.xlabel ('Serum Cholesterol Level')
plt.ylabel ('Resting Blood Pressure')
plt.title ('Correlation between Cholesterol Level and Resting Blood Pressure')

#Note we have some extreme values, however, there seems to be a positive correlation between both variables. 
```

```python
sns.swarmplot(x='eia', y='mhra', data=df)
plt.xlabel('Exercise Induced Angina')
plt.ylabel('Maximum Heart Rate Achieved')
plt.title('Exercise Induced Angina and Maximum Heart Rate Achieved')
```

```python
sns.boxplot(x="new_dhd", y="mhra", data=df)
plt.xlabel ('Presence of Coronary Heart Blockage')
plt.ylabel ('Maximum Heart Rate Achieved')
plt.title ('Maximum Heart Rate and Presence of Coronary Heart Blockage')
```

```python
sns.swarmplot(x='sex', y='rbp', data=df)
plt.xlabel('Gender')
plt.ylabel('Resting Blood Pressure')
plt.title('Gender and Resting Blood Pressure')
```

```python
sns.boxplot(x='new_dhd', y='rbp', data=df)
plt.xlabel('Presence of Coronary Heart Blockage')
plt.ylabel('Resting Blood Pressure')
plt.title('Presence of Coronary Heart Blockage and Resting Blood Pressure')
```

```python
sns.boxplot(x='cp', y='mhra', data=df)
plt.xlabel('Chest Pain Type')
plt.ylabel('Maximum Heart Rate Achieved')
plt.title('Presence of Coronary Heart Blockage and Chest Pain Type')
```

```python
sns.swarmplot(x='cp', y='rbp', data=df)
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Pressure')
plt.title('Chest Pain Type and Resting Blood Pressure')
```

```python
sns.swarmplot(x='rer', y='rbp', data=df)
plt.xlabel('Resting Electrocardiographic Results')
plt.ylabel('Resting Blood Pressure')
plt.title('Resting Electrocardiographic Results and Resting Blood Pressure')
```

```python
#Slicing dataset by gender
male= df[df.sex==1]
female= df[df.sex!=1]
```

```python
male.rbp.describe()
```

```python
#Histogram Plot for males and females resting blood pressure

plt.subplot(2,1,1)
plt.hist(male.rbp.dropna(), color='blue')
plt.title('Resting Blood Pressure by Gender')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Distribution')
plt.subplot(2,1,2)
plt.hist(female.rbp.dropna(), color='red')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Distribution')
plt.tight_layout()
plt.show()

```

```python
#Histogram plot for males and female serum cholesterol levels
plt.subplot(2,1,1)
plt.hist(male.scl.dropna(), color='blue')
plt.title('Serum Cholesterol Level by Gender')
plt.xlabel('Serum Cholesterol Level')
plt.ylabel('Distribution')
plt.subplot(2,1,2)
plt.hist(female.scl.dropna(), color='red')
plt.xlabel('Serum Cholesterol Level')
plt.ylabel('Distribution')
plt.tight_layout()
plt.show()
```
