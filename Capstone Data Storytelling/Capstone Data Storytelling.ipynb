```python
#Importing Pandas and Matplotlib and Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#Assigning column names 
col_names = ['age', 'sex', 'cp', 'rbp', 'scl', 'fbs', 'rer', 'mhra', 'eia', 'std', 'stss', 'fcv', 'tls', 'dhd']
```

```python
#Reading in the Cleveland Dataset 
data = pd.read_csv('Files/processed.cleveland.data.csv', header=None, names=col_names)
```

```python
#Switzerland Dataset
data1 = pd.read_csv('Files/processed.switzerland.data.csv', header=None, names=col_names)
```

```python
#Long Beach Dataset 
data2 = pd.read_csv('Files/processed.va.data.csv', header=None, names=col_names)
```

```python
#Hungarian Dataset 
data3 = pd.read_csv('Files/reprocessed.hungarian.data.csv', sep=" ", header=None, names=col_names)
```

```python
#Appending all dataset 
dataset = (data.append([data1, data2, data3],sort=True))
```

```python
#Identifying dataset shape
print(dataset.shape)
```

```python
#Printing dataset head
print(dataset.head())
```

```python
#Printing datset tail
print(dataset.tail())
```

```python
#Resetting the index for the dataset
dataset.reset_index()
```

```python
#Printing datset info
dataset.info()
```

```python
#Identifying whether there are any null values
dataset.isnull().values.any()
```

```python
#Dropping null values in dataset
dataset.dropna(axis=0, how='any', inplace=True)
#Removing -9 rows 
dataset=dataset[dataset != -9]
#Removing '?' rows
dataset=dataset[dataset != '?']
#Assigning all types to 'float'
dataset=dataset.astype(float)
```

```python
dataset.info()
```

```python
#Identifying age distribution of patients
dataset.plot(y='age', kind='hist', bins=15)
plt.xlabel('Patient Age Distribution')
plt.show()
```

```python
#Scatter plot studying age and resting blood pressure
dataset.plot(x='age', y='rbp', kind='scatter')
plt.xlabel('Patient Age')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
#Studying the relationship between sex and resting blood pressure
sns.swarmplot(x='sex', y='rbp', data=dataset)
```

```python
#studying the relationship between sex and serum cholestrol levels
sns.boxplot(x='sex', y='scl', data=dataset)
```

```python
#studying exercise induced angina by gender
sns.barplot(x='sex', y='eia', data=dataset)
```

```python
sns.barplot(x='sex', y='dhd', data=dataset)
```

```python
dataset.plot(x='dhd', y='rbp', kind='scatter')
plt.xlabel('% of blocked major blood vessels')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
dataset.plot(x='scl', y='rbp', kind='scatter')
plt.xlabel('Serum Cholesterol Level')
plt.ylabel('Resting Blood Pressure')
plt.show()
```
